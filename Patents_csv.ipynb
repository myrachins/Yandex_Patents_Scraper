```python
csv_features_file_name = 'features_full.csv'
json_features_directory = 'json_features/'
```

```python
_, _, json_features_file_names = next(walk(json_features_directory))
```

```python
columns = set()

for json_features_file_name in json_features_file_names:
    with open(json_features_directory + json_features_file_name, 'r') as json_file:
        for line in json_file:
            json = eval(line.strip())
            for key in json:
                columns.add(key)
            
columns = list(columns)
print(columns)
```

```python
with open(csv_features_file_name, 'w', encoding='utf-8') as f:
    f.write('\t'.join(columns))
    f.write('\n')
```

```python
for json_features_file_name in json_features_file_names:
    with open(json_features_directory + json_features_file_name, 'r') as json_file, \
         open(csv_features_file_name, 'a', encoding='utf-8') as csv_file:
        for line in json_file:
            json = eval(line.strip())
            for i, column in enumerate(columns):
                if i != 0:
                    csv_file.write('\t')
                if column in json and json[column] is not None:
                    csv_file.write(str(json[column]))
            csv_file.write('\n')
```

```python
def convert_to_list(row):
    try:
        row_eval = eval(row)
        if type(row_eval) is list:
            return row_eval
    except:
        return [row] if not pd.isna(row) else row
```

```python
def add_years(d, years):
    try:
        return d.replace(year = d.year + years)
    except ValueError:
        return d + (datetime.date(d.year + years, 1, 1) - datetime.date(d.year, 1, 1))
```

```python
def update_data_with_citations(data, author_citation_data, k):
    good_data = (pd.merge(author_citation_data, data[['patent_id', 'request_date']],
                        left_on='citated_by', right_on='patent_id', how='inner'))

    good_data['max_date'] = good_data['request_date_x'].apply(lambda date: add_years(date, k))
    good_data = good_data[good_data['request_date_y'] <= good_data['max_date']]
    
    citations_count = (good_data
                      .groupby('patent_id_x')
                      .agg({'citated_by': 'count'})
                      .rename(columns={'citated_by': f'yandex_cited_by_{k}'}))

    return (pd.merge(data, citations_count,
            left_on='patent_id', right_on='patent_id_x', how='left')
           .fillna({f'yandex_cited_by_{k}': 0}))
```

```python
def find_prod(data, patent_ids, k=None):
    column_name = ('yandex_cited_by_number' if k is None
                   else f'yandex_cited_by_{k}')
    return sum(data.loc[patent_ids][column_name])
```

```python
def create_prod_data(author_patent_data, year, k=None):
    column_name = (f'prod_{year}' if k is None
                   else f'prod_{year}_k{k}')
    return (author_patent_data
           .query(f'year == {year}')
           .groupby('author')
           .agg({'patent_id': lambda x: find_prod(data, list(x), k)})
           .rename(columns={'patent_id': column_name}))
```

```python
def get_prod6(authors_data, authors, year):
    return list(authors_data.loc[authors][f'cum_prod_{year}_k{6}'])
```

```python
prod_tmp_data = (author_patent_data
                 .groupby('patent_id')
                 .agg({'author': list, 'year': min}))

prod_tmp_data['prev_year'] = prod_tmp_data['year'] - 1
prod_tmp_data['cum_group_prod6_prev_year'] = (prod_tmp_data
                                       .apply(lambda row: get_prod6(authors_data, row['author'], row['prev_year']), axis=1))

prod_tmp_data.head()
```

```python
def find_int(row, high=True):
    prev_patent_year = row['prev_year']
    current_prod = row[f'cum_prod_{prev_patent_year}_k{6}']
    group_prods = row['cum_group_prod6_prev_year']
    
    return (sum(current_prod <= prod for prod in group_prods) if high
           else sum(current_prod > prod for prod in group_prods))
```

```python
merged_patent_author = (pd.merge(author_patent_data,
                                 prod_tmp_data[['cum_group_prod6_prev_year', 'prev_year']],
                                 on='patent_id', how='left'))

merged_patent_author = pd.merge(merged_patent_author, authors_data, on='author', how='left')
merged_patent_author['patent_high_int'] = (merged_patent_author
                                          .apply(lambda row: find_int(row, high=True), axis=1))
merged_patent_author['patent_low_int'] = (merged_patent_author
                                          .apply(lambda row: find_int(row, high=False), axis=1))

merged_patent_author[['author', 'patent_id', 'patent_high_int', 'patent_low_int', 'year']].head()
```

```python
interactions_data = (merged_patent_author
                    .groupby(['author', 'year'])
                    .agg({'patent_high_int': sum, 'patent_low_int':sum}))

interactions_data.head()
```

```python
def add_author_int(authors_data, interactions_data, year, high=True):
    int_type = 'high' if high else 'low'
    authors_data[f'{int_type}_int_current_{year}'] = (authors_data['author']
                                        .apply(lambda author: interactions_data.loc[(author, year)][f'patent_{int_type}_int']
                                                              if (author, year) in interactions_data.index
                                                              else 0))
    if high:
        authors_data[f'high_int_current_{year}'] -= authors_data[f'patents_number_{year}']
    
    authors_data[f'{int_type}_int_{year}'] = sum([authors_data[f'{int_type}_int_current_{y}']
                                                  for y in range(start_date, year+1)])
```

```python
for year in tqdm(range(start_date, end_date+1)):
    add_author_int(authors_data, interactions_data, year, high=True)
    add_author_int(authors_data, interactions_data, year, high=False)
    
authors_data.head()
```

```python
def get_leader(authors_data, authors, year):
    cum_prods = authors_data.loc[authors][f'cum_prod_{year-1}_k{6}']
    return np.argmax(cum_prods)
```

```python
authors_data.head()
```

```python
data['leader'] = data[['authors', 'year']].progress_apply(lambda row:
                                                          get_leader(authors_data, row['authors'], int(row['year'])), axis=1)

data.head()
```

```python
def get_int_leader(authors_data, row, high=True):
    prev_year = int(row['year']) - 1
    inter_type = 'high' if high else 'low'
    return authors_data.loc[row['leader']][f'{inter_type}_int_{prev_year}']
```
